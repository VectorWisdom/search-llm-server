```python
import utils as utl
import tiktoken
import numpy as np
import matplotlib.pyplot as plt
```

```python
elements = utl.load_json("../.data/pages_elements.json")
```

```python
print(len(elements))
```

```python
encoding = tiktoken.encoding_for_model("text-embedding-ada-002")
total = 0
for el in elements:
    el["token_len"] = len(encoding.encode(el["element_value"]))
    total += el["token_len"]
print(f"total of {total} tokens")
estimated_cost = (total / 1000)*0.0001
print(f"estimated cost is $ {estimated_cost}")
```

```python
counts = [el["token_len"] for el in elements if(el["token_len"]>2)]
```

```python
plt.hist(counts,bins=100)
plt.xlabel('Count')
plt.ylabel('Frequency')
plt.title('Histogram of Counts')
plt.show()
```
