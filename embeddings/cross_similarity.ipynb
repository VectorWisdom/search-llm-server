```python
import utils as utl
import numpy as np
from sklearn.preprocessing import normalize
from similarity import compute_similarity_matrix
import matplotlib.pyplot as plt
```

```python
def plot_similarity_heatmap(keys, similarity_matrix):
    plt.figure(figsize=(10, 8))
    plt.imshow(similarity_matrix, cmap='hot', interpolation='nearest')
    plt.colorbar()
    plt.xticks(ticks=range(len(keys)), labels=keys, rotation=90)
    plt.yticks(ticks=range(len(keys)), labels=keys)
    plt.title("Vector Similarity Heatmap")
    plt.show()
```

```python
def get_first_n_items(d, n):
    return {k: d[k] for k in list(d)[:n]}
```

```python
vectors_file="../.data/vectors.pkl"
# 'text-embedding-ada-002', 'all-MiniLM-L6-v2'
model="all-MiniLM-L6-v2"

embeddings = utl.load_pickle(vectors_file)
vectors = embeddings[model]
print(f" pickle file returned {len(vectors)} vectors for model '{model}'")
```

# all vectors all-MiniLM-L6-v2

```python
keys, similarity_matrix = compute_similarity_matrix(vectors)
plot_similarity_heatmap(keys, similarity_matrix)
```

# 100 vectors all-MiniLM-L6-v2

```python
few_vectors =  get_first_n_items(vectors,100)
keys, similarity_matrix = compute_similarity_matrix(few_vectors)
plot_similarity_heatmap(keys, similarity_matrix)
```
